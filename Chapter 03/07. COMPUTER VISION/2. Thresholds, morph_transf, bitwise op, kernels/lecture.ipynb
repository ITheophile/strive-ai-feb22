```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

# Bitwise operations

```python
canva = np.zeros((400, 400))
canva1, canva2 = canva.copy(), canva.copy()

# Show canva
plt.figure(figsize=(8, 8))
plt.imshow(canva, cmap='gray');
```

```python
# Draw a square on canva1
cv.rectangle(canva1, (50, 50), (250, 250), 255, cv.FILLED)
plt.figure(figsize=(8, 8))
plt.imshow(canva1, cmap='gray');
```

```python
# Draw a circle on canva2
cv.circle(canva2, (200, 200), 100, 255, cv.FILLED)
plt.figure(figsize=(8, 8))
plt.imshow(canva2, cmap='gray');
```

## And operator.  
It creates an **intersection**.

```python
and_op = cv.bitwise_and(canva1, canva2)

fig, axs = plt.subplots(1, 3, sharey=True, figsize = (15, 15))
axs[0].imshow(canva1, cmap='gray')
axs[1].imshow(canva2, cmap='gray')
axs[2].imshow(and_op, cmap ='gray')
plt.tight_layout();
```

## Or operator
It creates a **union**

```python
or_op = cv.bitwise_or(canva1, canva2)

fig, axs = plt.subplots(1, 3, sharey=True, figsize = (15, 15))
axs[0].imshow(canva1, cmap='gray')
axs[1].imshow(canva2, cmap='gray')
axs[2].imshow(or_op, cmap ='gray')
plt.tight_layout();
```

## Xor operator
It reverses the intersected area (**Not Intersection**)

```python
xor_op = cv.bitwise_xor(canva1, canva2)

fig, axs = plt.subplots(1, 3, sharey=True, figsize = (15, 15))
axs[0].imshow(canva1, cmap='gray')
axs[1].imshow(canva2, cmap='gray')
axs[2].imshow(xor_op, cmap ='gray')
plt.tight_layout();
```

## Not
It produces the **opposite**.

```python
not_and_op = cv.bitwise_not(and_op)
not_or_op = cv.bitwise_not(or_op)
not_xor_op = cv.bitwise_not(xor_op)

fig, axs = plt.subplots(1, 3, sharey=True, figsize = (15, 15))
axs[0].imshow(not_and_op, cmap='gray')
axs[1].imshow(not_or_op, cmap='gray')
axs[2].imshow(not_xor_op, cmap ='gray')
plt.tight_layout();
```

# Canny Detection

```python
img = cv.imread('images/day3.jpg', cv.IMREAD_GRAYSCALE)
blured_img = cv.GaussianBlur(img, (7, 7), 0)
canny_img = cv.Canny(img, 200, 400)
blured_canny_img = cv.Canny(blured_img, 125, 150)

fig, axs = plt.subplots(1, 3, sharey=True, figsize = (15, 8))
axs[0].imshow(img, cmap = 'gray')
axs[1].imshow(canny_img, cmap = 'gray')
axs[2].imshow(blured_canny_img, cmap = 'gray')
plt.tight_layout();
```
