```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('images/monday.jpg')
rgb_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```
