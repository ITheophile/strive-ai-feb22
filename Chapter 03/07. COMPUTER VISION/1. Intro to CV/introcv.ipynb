```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
cv2.__version__
```

# Load image

```python
img = cv2.imread('images/18.jpg')
type(img), img.shape
```

# Show with cv2

```python
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Show with plt and color space convertion

```python
plt.figure(figsize=(15, 8))
plt.imshow(img);
```

```python
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(15, 8))
plt.imshow(rgb_img);
```

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
```

# Image Annotation

```python
img_copy = rgb_img.copy()
cv2.line(img_copy,(0, 0), (175, 100), (255, 0, 0), 2)
plt.figure(figsize=(15, 8))
plt.imshow(img_copy)
```
