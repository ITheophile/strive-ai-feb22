```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Load and show image
image = cv2.imread('images/m.jpg')
image.shape

rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
rgb_copy = rgb_img.copy()

plt.figure(figsize=(15, 8))
plt.imshow(rgb_img);
```

# Mask for viewing given colors

```python
hsv_img = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
# Only blues
B_lower_bound = (89, 30, 20 )
B_upper_bound = (125, 255, 255)
mask = cv2.inRange(hsv_img, B_lower_bound, B_upper_bound)

rgb_copy = rgb_img.copy()
rgb_copy[mask == 0] = [0, 0, 0]
plt.figure(figsize=(15, 8))
plt.imshow(rgb_copy);
```

```python
# Only greens
G_lower_bound = (40, 50, 20 )
G_upper_bound = (70, 255, 255)
mask = cv2.inRange(hsv_img, G_lower_bound, G_upper_bound)

rgb_copy = rgb_img.copy()
rgb_copy[mask == 0] = [0, 0, 0]
plt.figure(figsize=(15, 8))
plt.imshow(rgb_copy);
```

# Meme Creator

```python
mem_img = cv2.imread('images/money.jpg')
rgb_mm_img = cv2.cvtColor(mem_img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8, 15))
plt.imshow(rgb_mm_img)
```

```python
mm_copy = rgb_mm_img.copy()
font = cv2.FONT_HERSHEY_SIMPLEX
text1 = 'Hey Daddy'
text2 = 'Lemme give you some money'
size_text1 = cv2.getTextSize(text1, font, 8, 20)[0]
x_text1 = (mm_copy.shape[1] - size_text1[0])//2
y_text1 = (size_text1[1] + size_text1[1]//2)

size_text1_ = cv2.getTextSize(text1, font, 8, 10)[0]
x_text1_ = (mm_copy.shape[1] - size_text1_[0])//2
y_text1_ = (size_text1_[1] + size_text1[1]//2)

size_text2 = cv2.getTextSize(text2, font, 6, 20)[0]
x_text2 = (mm_copy.shape[1] - size_text2[0])//2
y_text2 = (mm_copy.shape[0] - size_text2[1])

size_text2_ = cv2.getTextSize(text2, font, 6, 10)[0]
x_text2_ = (mm_copy.shape[1] - size_text2_[0])//2
y_text2_ = (mm_copy.shape[0] - size_text2_[1])


cv2.putText(mm_copy, 'Hey Daddy!!!', (x_text1, y_text1), cv2.FONT_HERSHEY_SIMPLEX, 8, (0, 0, 0), 20)
cv2.putText(mm_copy, 'Hey Daddy!!!', (x_text1_, y_text1_), cv2.FONT_HERSHEY_SIMPLEX, 8, (255, 255, 255), 10)

cv2.putText(mm_copy, 'Lemme give you some money', (x_text2, y_text2), cv2.FONT_HERSHEY_SIMPLEX, 6, (0, 0, 0), 20)
cv2.putText(mm_copy, 'Lemme give you some money', (x_text2_, y_text2_), cv2.FONT_HERSHEY_SIMPLEX, 6, (255, 255, 255), 10) 
dim_x = int((mm_copy.shape[1] * 0.25)/2)
dim_y = int((mm_copy.shape[0] * 0.25)/2)

mm_copy = cv2.resize(mm_copy, (dim_x, dim_y))
cv2.imwrite('images/meme.png', cv2.cvtColor(mm_copy, cv2.COLOR_RGB2BGR))

plt.figure(figsize=(8, 15))
plt.imshow(mm_copy)
```

# QR Code Reader

```python
qr = cv2.imread('images/qr_code.png')
qr_copy = qr.copy()
plt.imshow(qr)
```

```python
def qr_code_reader(filename):
    
    try:
        img = cv2.imread(filename)
        
        decoder = cv2.QRCodeDetector()
        text, bbox, straitqr = decoder.detectAndDecode(img)
        if bbox is not None:
            print('QR Code Detected')
            print(f'Encoded text : {text}')
            bbox = bbox.squeeze().astype(int)
            for i in range(len(bbox)):
                next_pt_idx = (i + 1) % len(bbox)
                cv2.line(img, tuple(bbox[i]), tuple(bbox[next_pt_idx]), (0, 0, 255), 2)
            plt.figure(figsize=(8, 8))
            plt.imshow(img)
            plt.show()
    except:
        print('QR Code Not Detected!!!')



qr_code_reader('images/qr_code.png')
```
